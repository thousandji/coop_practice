```python
import pymysql
```

```python
#mysql workbench에 practice라는 schema를 추가로 생성하고, task_data를 미리 업로드히였습니다

db = pymysql.connect(host='localhost', port=3306, user='root', passwd='1119', db='practice')
print(db) 
```

```python
cursor = db.cursor()
print(cursor)
```

```python
#task_data를 전체 조회하기 위한 쿼리문을 작성하여 파이썬으로 실행하는 코드입니다
sql = "SELECT * FROM task_data"
cursor.execute(sql)

result = cursor.fetchall()
for record in result:
    print(record)
db.close()
```
